```python
import math

def generate_perms(n, m):
    """
    Generate configurations of S_n with a maximum cycle length of m.
    
    Args:
    - n (int): The total number to be partitioned.
    - m (int): The maximum cycle length.
    
    Returns:
    - list: A list of configurations.
    """
    configs = []
    
    if n == 0:
        return [[]]
    elif n > 0 and m == 0:
        return None
    else: 
        for i in range(1, min(m+1, n+1)):
            L = generate_perms(n-i, i)
            if L:
                for l in L:
                    l.append(i)
                    configs.append(l)
    return configs

def num_perms(n, configs):
    """
    Calculate the number of permutations for a given configuration.
    
    Args:
    - n (int): The total number.
    - configs (list): The configuration list.
    
    Returns:
    - int: The number of permutations.
    """
    result = math.factorial(n)
    p = 1
    for k in range(1, n+1):
        a_k = configs.count(k)
        p *= math.factorial(a_k) * math.pow(k, a_k)
    return int(result // p)

def gcd(x, y):
    """
    Calculate the Greatest Common Divisor (GCD) of two numbers.
    
    Args:
    - x (int): First number.
    - y (int): Second number.
    
    Returns:
    - int: The GCD of x and y.
    """
    if y == 0:
        return x
    return gcd(y, x % y)

def fix(p, q, s):
    """
    Calculate the number of invariant states for given partitions.
    
    Args:
    - p (list): Partition for width.
    - q (list): Partition for height.
    - s (int): Possible states for each celestial body.
    
    Returns:
    - int: The number of invariant states.
    """
    return pow(s, sum([sum([gcd(row, col) for row in p]) for col in q]))

def solution(w, h, s):
    """
    Calculate the number of unique configurations on a star grid.
    
    Args:
    - w (int): Width of the grid.
    - h (int): Height of the grid.
    - s (int): Possible states for each celestial body.
    
    Returns:
    - str: The number of unique configurations.
    """
    P = generate_perms(w, w)
    Q = generate_perms(h, h)
    result = 0
    for p in P:
        for q in Q:
            result += fix(p, q, s) * num_perms(w, p) * num_perms(h, q)
    result = result // (math.factorial(w) * math.factorial(h))
    return str(int(result))

# Test the solution
print(solution(2, 2, 2))
```

```python
from math import factorial, gcd
from itertools import permutations

def cycle_count(p):
    """Count the cycles in the permutation p."""
    remaining = set(p)
    cycles = 0
    while remaining:
        i = remaining.pop()
        cycles += 1
        j = p[i]
        while j != i:
            remaining.remove(j)
            j = p[j]
    return cycles

def solution(w, h, s):
    """Calculate the number of unique, non-equivalent configurations."""
    count = 0
    for p in permutations(range(w)):
        for q in permutations(range(h)):
            cycles = cycle_count(p) + cycle_count(q)
            count += pow(s, cycles)

    # Now divide by the number of permutations of rows and columns to get the final answer.
    return str(count // (factorial(w) * factorial(h)))

# Test with some sample cases
print(solution(2, 2, 2))  # Expected: 7
print(solution(2, 3, 4))  # Expected: 430
```
